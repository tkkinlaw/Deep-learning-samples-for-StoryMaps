## Using custom transforms in model training

### Import libaries

```python
from arcgis.gis import GIS
from arcgis.learn import prepare_data, MaskRCNN
```

```python
gis = GIS("home")
```

### Prepare for the transforms

The `transforms` parameter in `prepare_data` expects a tuple or list containing the tranforms for the training data and the validation data.

```python
from fastai.vision.transform import rotate, contrast, crop
```

```python
trainingTfms = [rotate(degrees=30, p=1), 
                contrast(scale=[1.75, 2], p=1), 
                crop(size=150, row_pct=0.3, col_pct=0.6, p=1)]
validationTfms = [crop(size=150, row_pct=0.3, col_pct=0.6, p=1)]
tfms = (trainingTfms, validationTfms)
```

### Prepare the image chips for training

```python
dataPath = r'/arcgis/home/Building Detection/DLTrainingSamples'
```

```python
data = prepare_data(dataPath, batch_size=16, chip_size=112, class_mapping={'1': 'Building'}, transforms=tfms)
```

Here you can see that some images are transformed.

```python
data.show_batch()
```

### Train the MaskRCNN model

```python
model = MaskRCNN(data, backbone='resnet50')
```

```python
lr = model.lr_find()
lr.real
```

```python
model.fit(epochs=20, lr=lr.real)
```

```python
model.plot_losses()
```

```python
model.average_precision_score()
```
